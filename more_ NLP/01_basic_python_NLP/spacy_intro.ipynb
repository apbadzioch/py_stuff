spaCy is used in NLP

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
# create an object
doc = nlp("I am going to learn more about NLP, LLMs, and how to RAGs and Agents.")

# print each token separately
# pos_ = part of speech
# dep_ = dependency
for token in doc:
    print(token.text, token.pos_, token.dep_)
```
