# Tokenization
The first step in creating a DOC object is to break down the incoming text into component pieces called "tokens".

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
# creating a sentence that includes quotations.
mystring = '"We\'re going to learn more about Natural Language Processing."'
print(mystring)
```

```python
# create the object and explore strings
doc = nlp(mystring)
for token in doc:
    print(token.text, end=' | ')
```
